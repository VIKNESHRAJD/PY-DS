# 1. Get user input and display it


```python
name = input("Enter your name: ")
age = input("Enter your age: ")
email = input("Enter your email: ")
phone = input("Enter your phone number: ")
print("Name:", name)
print("Age:", age)
print("Email:", email)
print("Phone:", phone)
```

## 2. Operations on two integers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Sum:", a + b)
print("Difference:", a - b)
print("Product:", a * b)
print("Quotient:", a // b)
print("Remainder:", a % b)
```

```python
# 3. Celsius to Fahrenheit and vice versa
c = float(input("Enter temperature in Celsius: "))
f = (1.8 * c) + 32
print("Fahrenheit:", f)
f = float(input("Enter temperature in Fahrenheit: "))
c = (f - 32) / 1.8
print("Celsius:", c)
```

```python
# 4. Kilometer to Mile and vice versa
km = float(input("Enter distance in kilometers: "))
mile = 0.621371 * km
print("Miles:", mile)
mile = float(input("Enter distance in miles: "))
km = mile / 0.621371
print("Kilometers:", km)
```

```python
# 5. Square root
import math
num = float(input("Enter a number: "))
print("Square root:", math.sqrt(num))
```

```python
# 6. Area of triangle using Heron’s formula
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))
s = (a + b + c) / 2
area = math.sqrt(s * (s - a) * (s - b) * (s - c))
print("Area of triangle:", area)
```

```python
# 7. Swap two variables
x = input("Enter first value: ")
y = input("Enter second value: ")
x, y = y, x
print("After swapping: x =", x, ", y =", y)
```

```python
# 8. Generate random number
import random
print("Random number:", random.randint(1, 100))
```

```python
# 5. Simple Interest
P = float(input("Enter principal (P): "))
N = float(input("Enter time in years (N): "))
R = float(input("Enter rate of interest (R): "))
SI = (P * N * R) / 100
print("Simple Interest:", SI)
```

```python
# 6. Compound Interest
P = float(input("Enter principal (P): "))
R = float(input("Enter rate of interest (R): "))
n = float(input("Enter time in years (n): "))
CI = P * ((1 + R / 100) ** n)
print("Compound Interest:", CI)
```
